```python
import numpy as np
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

- Простые линейные графики

```python
# 1-мерный массив
x_arr_0 = np.arange(10)
```

```python
x_arr_0
```

```python
y_arr_0 = x_arr_0

# визуализация
plt.plot(x_arr_0, y_arr_0)

plt.show()
```

```python
# совмещение нескольких графиков в одном окне
y_arr_1 = x_arr_0 ** 2
y_arr_2 = x_arr_0 ** 3

plt.plot(x_arr_0, y_arr_1)
plt.plot(x_arr_0, y_arr_2)

plt.show()
```

```python
#  1-мерный массив из определенного количества значений (100 значений)
# В диапазоне от -10 до 10
x_arr_1 = np.linspace(-10, 10, 100)
x_arr_1
```

```python
# функция синуса
y_1 = np.sin(x_arr_1)

# функция гиперболического тангенса
y_2 = np.tanh(x_arr_1)

# функция сигмоиды
def sigmoid(x):
    return 1 / (1 +  np.exp(-x))

y_3 = sigmoid(x_arr_1)

plt.plot(x_arr_1, y_1)
plt.plot(x_arr_1, y_2)
plt.plot(x_arr_1, y_3)

plt.show()
```

```python
# цвета линий

y_1 = np.sin(x_arr_1)
y_2 = np.sin(x_arr_1 + 1)
y_3 = np.sin(x_arr_1 + 2)
y_4 = np.sin(x_arr_1 + 3)

# по названию цвета
plt.plot(x_arr_1, y_1, color="red")
# по первой букве названия цвета из формата RGB и CMYK
plt.plot(x_arr_1, y_2, c="m")
# по 16 кодировке RGB или RGBA
plt.plot(x_arr_1, y_3, c="#341fa8")
# по формату RGB или RGBA в виде кортежа со значениями от 0 до 1
plt.plot(x_arr_1, y_4, c=(0.0, 0.8, 0.2, 1.0))
plt.show()
```

```python
# стили линий

# сплошная линия 
plt.plot(x_arr_1, y_1, linestyle="solid")
# штриховая
plt.plot(x_arr_1, y_2, ls="dashed")
# пунктирная
plt.plot(x_arr_1, y_3, ls="dotted", c="#0afaa2", lw=4)
# штрих-пунктриная линия
plt.plot(x_arr_1, y_4, ls="dashdot", c="#faff00", lw=4)
```

```python
# обьединение настроек 

# штриховая зеленая
plt.plot(x_arr_1, y_1, "--g")
# пунктриная синяя
plt.plot(x_arr_1, y_2, ":b")
# штрих пунктирная желтая
plt.plot(x_arr_1, y_3, "-.y")

# "звездная" красная
plt.plot(x_arr_1, y_4, "*r")
```

```python
# заголовок, подписи осей, легенда графика

# размер окна графика
plt.figure(figsize=(10,8))

plt.plot(x_arr_1, np.sin(x_arr_1), ":r")
plt.plot(x_arr_1, np.cos(x_arr_1), "+g")
# заголовок
plt.title("Графики синуса и косинуса")
# подписи осей
plt.xlabel("ось x")
plt.ylabel("ось y")
# легенда графика
plt.legend(["синус", "косинус"])

# диапазон коордитнатной сетки
plt.axis([-15, 15, -2, 2])
plt.show()

```
